```python
import json
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # this import registers the 3D projection, even if unused

# Load overall results from JSON
with open("overall_results_2.json", "r") as f:
    overall_results = json.load(f)

gammas = []
kappas = []
returns = []  # here we use average cash as the "return"

# Iterate over overall_results (assumes keys like "gamma_kappa")
for key, val in overall_results.items():
    # For example, each val might have "avg_cash", "gamma" and "kappa"
    gammas.append(val["gamma"])
    kappas.append(val["kappa"])
    returns.append(val["avg_cash"])

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(gammas, kappas, returns, c=returns, cmap='viridis', marker='o')

ax.set_xlabel("Gamma")
ax.set_ylabel("Kappa")
ax.set_zlabel("Average Cash Return")
ax.set_title("Overall Returns vs Gamma and Kappa")

# Optionally add a colorbar to see how color maps to returns

```

```python
import json
import plotly.express as px

# Load overall results from JSON file
with open("overall_results_2.json", "r") as f:
    overall_results = json.load(f)

# Create lists (or you could create a list of dicts)
data = []
for key, val in overall_results.items():
    data.append({
        "gamma": val["gamma"],
        "kappa": val["kappa"],
        "avg_cash": val["avg_cash"]
    })

# Create an interactive 3D scatter plot using Plotly Express.
fig = px.scatter_3d(data, x="gamma", y="kappa", z="avg_cash",
                      color="avg_cash", color_continuous_scale="Viridis",
                      title="Interactive 3D Plot of Returns vs Gamma and Kappa",
                      labels={"gamma": "Gamma", "kappa": "Kappa", "avg_cash": "Average Cash Return"})

fig.show()
```
